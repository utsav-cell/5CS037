Problem 2: Data Imputations
**Dataset:** `medical_student.csv`


```python

medical_df = pd.read_csv('medical_student.csv')
medical_df.head()
```

```python

medical_df.info()
print('\nMissing values per column:')
print(medical_df.isnull().sum())
```

```python

for col in medical_df.columns:
    if medical_df[col].dtype == 'object':
        medical_df[col].fillna(medical_df[col].mode()[0], inplace=True)
    else:
        medical_df[col].fillna(medical_df[col].mean(), inplace=True)

print('Filled missing values successfully!')
medical_df.isnull().sum()
```

```python

print('Duplicate rows:', medical_df.duplicated().sum())
medical_df = medical_df.drop_duplicates()
print('After removing duplicates:', medical_df.duplicated().sum())
```
