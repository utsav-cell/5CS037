 Problem 3: Array Operations

```python
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

print('Add:')
print(x + y)
print('Subtract:')
print(x - y)
print('Multiply by 2:')
print(x * 2)
print('Square:')
print(x ** 2)
print('Dot v.w:', np.dot(v,w))
print('Dot x.v:', np.dot(x,v))
print('Dot x.y:\n', np.dot(x,y))
print('Concatenate x and y:\n', np.concatenate((x,y), axis=0))
print('Concatenate v and w:\n', np.vstack((v,w)))
try:
    print(np.concatenate((x,v)))
except Exception as e:
    print('Error combining x and v:', e)
```

 Problem 4: Matrix Operations

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print('A * A^-1:')
print(I)
AB = np.dot(A,B)
BA = np.dot(B,A)
print('AB:\n', AB)
print('BA:\n', BA)
lhs = AB.T
rhs = np.dot(B.T, A.T)
print('(AB)^T:\n', lhs)
print('B^T * A^T:\n', rhs)
A_eq = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_eq = np.array([-1, -3, 9])
sol = np.linalg.solve(A_eq, B_eq)
print('Solution using solve:', sol)
A_inv_eq = np.linalg.inv(A_eq)
sol2 = np.dot(A_inv_eq, B_eq)
print('Solution using inverse:', sol2)
```
